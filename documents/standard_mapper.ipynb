Here are some exercises on conventional mapper. 

```python
#In order to instal keppler mapper (https://kepler-mapper.scikit-tda.org/) please run 
import kmapper as km
from kmapper import jupyter
import matplotlib.pyplot as plt
import kmapper as km
from sklearn import datasets
import networkx as nx
```

The initial exercise is motivated by:
https://kepler-mapper.scikit-tda.org/notebooks/KeplerMapper-usage-in-Jupyter-Notebook.html

```python
data, labels = datasets.make_circles(n_samples=5000, noise=0.03, factor=0.3)
plt.scatter(data[:,0],data[:,1])
plt.show()
data = 10*data
```

Now, let us do some mapper!

```python
mapper = km.KeplerMapper()
lens = data[:,1] + data[:,0]
graph = mapper.map(X=data, lens=lens, nr_cubes = 10)

# Visualize it
html = mapper.visualize(graph, path_html="make_circles_keplermapper_output.html",
                 title="make_circles(n_samples=5000, noise=0.03, factor=0.3)")

jupyter.display(path_html="make_circles_keplermapper_output.html")
```

```python
#We can also convert it to a networx graph and plot it: 
nx_graph = km.adapter.to_nx(graph)
nx.draw(nx_graph)
```
